```python
#Eliminação de Gauss

import numpy as np

def ElimGauss(a,b):
 n = len(b)
 for k in range(0,n-1):
   for i in range(k+1,n):
     if a[i,k] != 0.0:
        lam = a[i,k]/a[k,k]
        a[i,k+1:n] = a[i,k+1:n] - lam*a[k,k+1:n]
        b[i] = b[i] - lam*b[k]        
 return a, b

def SolveSis(a,b):
 n = len(b)
 a, b = ElimGauss(a,b)
 for k in range(n-1,-1,-1):
    b[k] = (b[k] - a[k,k+1:n].dot(b[k+1:n]))/a[k,k]
 return b

#Exemplo 
A = np.array([[2, 1,2], [4, 3,3],[6, 5,-1]])
b = np.array([-1,0,6])
x = SolveSis(A,b)
print(x)
```
