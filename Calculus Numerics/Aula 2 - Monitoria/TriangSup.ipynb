```python
# Triangular Superior

def TriangularSup(A,b):
  n = len(b)
  x = [0]*n
  x[n-1] = b[n-1]/A[n-1][n-1]

  for i in list(range(n-1,0,-1)):
    s = 0
    for j in list(range(i+1,n+1,1)):
      s = s + A[i-1][j-1]*x[j-1]

    x[i-1] = (b[i-1]-s)/(A[i-1][i-1]) 

  return x
  
# Dados
A = [[3,2,4],[0,(1/3),(2/3)],[0,0,1]]
b = [1,(5/3),0]
x = TriangularSup(A,b)

print(x)
```
