```python
from math import sqrt
import numpy as np

def cholesky(A):
    n = np.shape(A)[0]

    L = np.zeros((n,n))

    for i in np.arange(n):
        for j in np.arange(i+1):
            tmp_sum = sum(L[i][k]*L[j][k] 
            for k in np.arange(j))           
            if (i == j):
                L[i][j] = sqrt(A[i][i] - tmp_sum)
            else:
                L[i][j] = ((A[i][j] - tmp_sum)/L[j][j])
    return L
 
A = [[3,2,0], [2,5,1], [0,1,3]]
L = cholesky(A)

print(L)
```
