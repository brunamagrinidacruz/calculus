```python
import numpy as np
from scipy import linalg as lin
import scipy as sci
import math as m
import matplotlib.pyplot as plt
from sympy import *
import time

def mmq(x,y,k):
    X = np.vander(x,k);
    A = np.transpose(X).dot(X);
    b = np.transpose(X).dot(y);
    a = np.linalg.solve(A,b);

    return a;

def mmq_qr(x,y,k):
    X = np.vander(x,k);
    
    (Q,R) = np.linalg.qr(X)
    
    b = np.transpose(Q).dot(y);
    a = np.linalg.solve(R,b);

    return a;

# Usando os dados do Covid coletados de 26 de Fevereiro 2020 até 06 de Julho 2021
totalcasos = np.genfromtxt(fname='drive/My Drive/covidcasosacumulados.txt') # y
dias = np.linspace(1, 499, num=499, endpoint=True) # x

# Plotar os dados
plt.figure(figsize=(15,6),facecolor='white')
plt.scatter(dias,totalcasos,label = 'Casos acumulados COVID-19 no Brasil',linewidth = 3,color='red')
# plt.xscale('log')
# plt.yscale('log')
plt.xlabel('x',fontsize='large') 
plt.ylabel('y',fontsize='large') 
plt.title('Observação dos casos de 26 de Fevereiro 2020 até 06 de Julho 2021') 
plt.legend(fontsize='large') 
plt.show()

# Aproximar para uma função ab^x, como estou aproximando por ln(ab^x) = ln(a) + ln(b)x, eu devo aplicar ln em total de casos
totalcasos_ln = np.log(totalcasos);
alpha = mmq_qr(dias,totalcasos_ln,2) # a aproximação é uma reta, logo k = 2
print(alpha);

# Transformando em a e b efetivamente
a = np.exp(alpha[1]); b = np.exp(alpha[0]);
print(a);
print(b);
g = lambda x: a*(b**x);

# Plotar os dados e a curva
plt.figure(figsize=(15,6),facecolor='white')
plt.plot(dias,g(dias),label = 'g(x)',linewidth = 2,marker='>',zorder=1)
plt.scatter(dias,totalcasos,label = 'Casos acumulados COVID-19 no Brasil',linewidth = 2,color='red',zorder=2)
plt.xlabel('x',fontsize='large') 
plt.ylabel('y',fontsize='large') 
plt.title('Observação dos casos de 26 de Fevereiro 2020 até 06 de Julho 2021') 
plt.legend(fontsize='large') 
plt.show()

# Eliminar os 20 primeiros dias
dias_479 = dias[-479:];
totalcasos_479 = totalcasos[-479:]

# Aproximar para uma função ab^x, como estou aproximando por ln(ab^x) = ln(a) + ln(b)x, eu devo aplicar ln em total de casos
totalcasos_479_ln = np.log(totalcasos_479);
alpha_479 = mmq_qr(dias_479,totalcasos_479_ln,2) # a aproximação é uma reta, logo k = 2
print(alpha_479);

# Transformando em a e b efetivamente
a_479 = np.exp(alpha_479[1]); b_479 = np.exp(alpha_479[0]);
print(a_479);
print(b_479);
g_479 = lambda x: a_479*(b_479**x);

# Plotar os dados e a curva
plt.figure(figsize=(15,6),facecolor='white')
plt.plot(dias_479,g_479(dias_479),label = 'g(x)',linewidth = 2,marker='>',zorder=1)
plt.scatter(dias_479,totalcasos_479,label = 'Casos acumulados COVID-19 no Brasil',linewidth = 2,color='red',zorder=2)
plt.xlabel('x',fontsize='large') 
plt.ylabel('y',fontsize='large') 
plt.title('Observação dos casos de 16 de Março 2020 até 06 de Julho 2021') 
plt.legend(fontsize='large') 
plt.show()

# Eliminar os 50 primeiros dias
dias_449 = dias[-449:];
totalcasos_449 = totalcasos[-449:]

# Aproximar para uma função ab^x, como estou aproximando por ln(ab^x) = ln(a) + ln(b)x, eu devo aplicar ln em total de casos
totalcasos_449_ln = np.log(totalcasos_449);
alpha_449 = mmq_qr(dias_449,totalcasos_449_ln,2) # a aproximação é uma reta, logo k = 2
print(alpha_449);

# Transformando em a e b efetivamente
a_449 = np.exp(alpha_449[1]); b_449 = np.exp(alpha_449[0]);
print(a_449);
print(b_449);
g_449 = lambda x: a_449*(b_449**x);

# Plotar os dados e a curva
plt.figure(figsize=(15,6),facecolor='white')
plt.plot(dias_449,g_449(dias_449),label = 'g(x)',linewidth = 2,marker='>',zorder=1)
plt.scatter(dias_449,totalcasos_449,label = 'Casos acumulados COVID-19 no Brasil',linewidth = 2,color='red',zorder=2)
plt.xlabel('x',fontsize='large') 
plt.ylabel('y',fontsize='large') 
plt.title('Observação dos casos de 16 de Abril 2020 até 06 de Julho 2021') 
plt.legend(fontsize='large') 
plt.show()

# Eliminar os 249 primeiros dias
dias_250 = dias[-250:];
totalcasos_250 = totalcasos[-250:]

# Aproximar para uma função ab^x, como estou aproximando por ln(ab^x) = ln(a) + ln(b)x, eu devo aplicar ln em total de casos
totalcasos_250_ln = np.log(totalcasos_250);
alpha_250 = mmq_qr(dias_250,totalcasos_250_ln,2) # a aproximação é uma reta, logo k = 2
print(alpha_250);

# Transformando em a e b efetivamente
a_250 = np.exp(alpha_250[1]); b_250 = np.exp(alpha_250[0]);
print(a_250);
print(b_250);
g_250 = lambda x: a_250*(b_250**x);

# Plotar os dados e a curva
plt.figure(figsize=(15,6),facecolor='white')
plt.plot(dias_250,g_250(dias_250),label = 'g(x)',linewidth = 2,marker='>',zorder=1)
plt.scatter(dias_250,totalcasos_250,label = 'Casos acumulados COVID-19 no Brasil',linewidth = 2,color='red',zorder=2)
plt.xlabel('x',fontsize='large') 
plt.ylabel('y',fontsize='large') 
plt.title('Observação dos casos - últimos 250 dias') 
plt.legend(fontsize='large') 
plt.show()

# Utilizar apenas os 20 primeiros dias
dias_20 = dias[:20];
totalcasos_20 = totalcasos[:20]

# Aproximar para uma função ab^x, como estou aproximando por ln(ab^x) = ln(a) + ln(b)x, eu devo aplicar ln em total de casos
totalcasos_20_ln = np.log(totalcasos_20);
alpha_20 = mmq_qr(dias_20,totalcasos_20_ln,2) # a aproximação é uma reta, logo k = 2
print(alpha_20);

# Transformando em a e b efetivamente
a_20 = np.exp(alpha_20[1]); b_20 = np.exp(alpha_20[0]);
print(a_20);
print(b_20);
g_20 = lambda x: a_20*(b_20**x);

# Plotar os dados e a curva
plt.figure(figsize=(15,6),facecolor='white')
plt.plot(dias_20,g_20(dias_20),label = 'g(x)',linewidth = 2,marker='>',zorder=1)
plt.scatter(dias_20,totalcasos_20,label = 'Casos acumulados COVID-19 no Brasil',linewidth = 2,color='red',zorder=2)
plt.xlabel('x',fontsize='large') 
plt.ylabel('y',fontsize='large') 
plt.title('Observação dos casos de 26 de Fevereiro 2020 até 15 de Março 2020') 
plt.legend(fontsize='large') 
plt.show()

print('\n')
```
